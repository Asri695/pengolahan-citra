```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

from google.colab.patches import cv2_imshow

ft = '/content/drives/MyDrive/pcd/gambar.jpg'

img = cv.imread (ft)
img1 = cv.cvtColor(cv.imread (ft), cv.COLOR_BGR2GRAY)
his = cv.equalizeHist(img1)

cv2_imshow(img)
cv2_imshow(img1)

plt.subplot (2,1,1),plt.hist(img.ravel(), bins = 256, range = [0,256])
plt.subplot (2,1,2),plt.hist(his.ravel(), bins = 256, range = [0,256])
plt.show()
```

```python
bright_image = (100.0/255)*img1 + 100
cv2_imshow(bright_image)

plt.subplot (2,1,1),plt.hist(his.ravel(), bins = 256, range = [0,256])
plt.subplot (2,1,2),plt.hist(bright_image.ravel(), bins = 256, range = [0,256])
plt.show()
```

```python
darken_image = 255.0*(img1/255.0)**2
cv2_imshow(darken_image)

plt.subplot (2,1,1),plt.hist(his.ravel(), bins = 256, range = [0,256])
plt.subplot (2,1,2),plt.hist(darken_image.ravel(), bins = 256, range = [0,256])
plt.show()
```
